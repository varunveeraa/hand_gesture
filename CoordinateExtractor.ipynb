```python
#importing libraries
import mediapipe as mp
import numpy as np
import cv2
```

```python
#mediapipe variables
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        image = cv2.flip(image, 1) #flip on horizontal
        
        image.flags.writeable = False #flag set false
        
        #detections
        results = hands.process(image)
        
        image.flags.writeable = True #flag set true
        
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        print(results)
        
        #rendering results
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)
            
        
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
#importing the libraries
import csv
import os
import numpy as np
```

```python
coordinates = len(results.multi_hand_landmarks[0].landmark)
```

```python
landmarks = ['class']
for val in range(1, coordinates+1):
    landmarks += ['x{}'.format(val), 'y{}'.format(val), 'z{}'.format(val)]
```

```python
with open('coordinates.csv', mode='w', newline='') as f:
    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(landmarks)
#here i've generated a csv with declared landmark variables, which will make it easy to append values
```

```python
class_name = "forward"
#if you want to add more gestures/actions, change the class_name after 
#running the below cell, and run the below cell again to append the 
#new gesture to csv file
```

```python
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        image = cv2.flip(image, 1) #flip on horizontal
        
        image.flags.writeable = False #flag set false
        
        #detections
        results = hands.process(image)
        
        image.flags.writeable = True #flag set true
        
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        print(results)
        
        #rendering results
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)
            
        #exporting the coordinates of the new gesture/action
        try:
            hand_pose = results.multi_hand_landmarks[0].landmark #
            rowTemp = list([[landmark.x, landmark.y, landmark.z] for landmark in hand_pose])
            row = sum(rowTemp, []) #addin' up lists
           
            #append class name 
            row.insert(0, class_name)
            
            #export to CSV
            with open('coordinates.csv', mode='a', newline='') as f:
                csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csv_writer.writerow(row) 
            
        except:
            pass
        
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
